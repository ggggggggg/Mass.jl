```julia
jldname = "/Volumes/Drobo/exafs_data/20141030_brown_ferrioxalate_1M_calibronium/20141030_brown_ferrioxalate_1M_calibronium_mass.hdf5"
jld = jldopen(jldname)
c=jld["chan1"]
```

```julia
steps = h5steps(c)
s=steps[1]
string(Mass.H5Flow.func(s))
Mass.H5Flow.p_inputs(s)
Mass.H5Flow.o_inputs(s)
Mass.H5Flow.p_outputs(s)
Mass.H5Flow.o_outputs(s)
```

```julia
label(v)=v.label
vertex_labels(g) = [label(v) for v in vertices(g)]
function add_exvertex!(g, name,shape)
    verts = vertices(g)
    vert_labels = [label(v) for v in verts]
    if !(name in vert_labels)
        v = ExVertex(num_vertices(g)+1, name)
        attrs = attributes(v,g)
        attrs["label"]=name
        attrs["shape"]=shape
        attrs["fontsize"]=18.0
        attrs["color"]="red"
        attrs["fillcolor"]="lightgrey"
        attrs["style"]="filled"
        add_vertex!(g, v)
        v
    else
        first(verts[indexin([name],vert_labels)])
    end
end
add_p_data!(g,name)=add_exvertex!(g,name,"doubleoctagon")
add_o_data!(g,name)=add_exvertex!(g,name,"octagon")
add_func!(g,name)=add_exvertex!(g,name,"box")
function Graphs.add_edge!(g,label1, label2)
    verts = vertices(g)
    vert_labels = [label(v) for v in verts]
    i1 = first(indexin([label1], vert_labels))
    i2 = first(indexin([label2], vert_labels))
    if i1 < 1 || i2 < 1
        error("error label $label1 or $label2 is not in $g")
    end
    add_edge!(g, verts[i1], verts[i2])
end
```

```julia
function add_step!(g, s)
    vf = add_func!(g,string(Mass.H5Flow.funcname(s)))
    for p in Mass.H5Flow.p_inputs(s)
        v=add_p_data!(g,p)
        add_edge!(g,v,vf)
    end
    for o in Mass.H5Flow.o_inputs(s)
        v=add_o_data!(g,o)
        add_edge!(g,v,vf)
    end
    for p in Mass.H5Flow.p_outputs(s)
        v=add_p_data!(g,p)
        add_edge!(g,vf,v)
    end
    for o in Mass.H5Flow.o_outputs(s)
        v=add_o_data!(g,o)
        add_edge!(g,vf,v)
    end
    g
end
```

```julia
g=inclist(ExVertex, is_directed=true)
for s in steps
    add_step!(g,s)
end
```

```julia
attrs = AttributeDict()
attrs["overlap"]="false"
attrs["overlap_shrink"]="true"
attrs["size"]="16,10"
gdot = to_dot(g, attrs)
gviz = GraphViz.Graph(gdot)
GraphViz.layout!(gviz,engine="dot")
gviz
```
